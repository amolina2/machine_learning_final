```python
import matplotlib.pyplot as plt
import torch

# Load the data for plotting
plot_data = torch.load('plot_data.pt')
train_losses = plot_data['train_losses']
val_losses = plot_data['val_losses']
actual_ages = plot_data['actual_ages']
predicted_ages = plot_data['predicted_ages']


plt.figure(figsize=(10, 6))
plt.plot(range(1, len(train_losses) + 1), train_losses, label='Training Loss')
plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss')
plt.title('Training and Validation Losses')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(range(len(actual_ages)), actual_ages, label='Actual Age', color='blue')
plt.scatter(range(len(predicted_ages)), predicted_ages, label='Predicted Age', color='red')
plt.title('Comparison of Actual and Predicted Ages -First 15 predictions')
plt.xlabel('Sample Index')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plot_data = torch.load('plot_data.pt')
actual_ages = plot_data['actual_ages']
predicted_ages = plot_data['predicted_ages']

df = pd.read_csv('UTKFaceAugmented.csv')

gender_info = df[['filename', 'gender']]

good_prediction_threshold = 5

good_predictions_male = 0
good_predictions_female = 0

for i in range(len(predicted_ages)):
   
    filename = df.iloc[i]['filename']
    gender = gender_info[gender_info['filename'] == filename]['gender'].values[0]

    
    is_good_prediction = abs(predicted_ages[i] - actual_ages[i]) <= good_prediction_threshold

    
    if gender == 'male' and is_good_prediction:
        good_predictions_male += 1
    elif gender == 'female' and is_good_prediction:
        good_predictions_female += 1

labels = ['Male', 'Female']
sizes = [good_predictions_male, good_predictions_female]
pie_colors = ['lightseagreen', 'salmon'] 


plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=pie_colors,startangle=140)
plt.title('Good Predictions by Gender')
plt.show()
```
